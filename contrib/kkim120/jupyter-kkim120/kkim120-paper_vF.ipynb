## Literature Review

The connection between inflation and stock prices has long been a subject of intense debate and analysis within the fields of finance and economics. Historical research reveals an intricate and often contentious relationship, rooted in economic theories and empirical observations. At the heart of this debate stands the Fisher Effect, a pivotal concept proposed by Irving Fisher in his 1930 work. Fisher's theory, while foundational, is marked by complexity and intricacies that underscore the multifaceted nature of this relationship. Irving Fisher's Fisher Effect theory stands as a cornerstone in understanding the interplay between inflation and stock prices. The Fisher Effect posits a positive correlation between nominal interest rates and inflation rates. According to this theory, as inflation increases, nominal interest rates tend to rise correspondingly. This phenomenon can have profound implications for various aspects of the economy, including the stock market. 


Fisher's argument hinges on the idea that investors and lenders, when confronted with inflation, will seek to protect the purchasing power of their investments. To achieve this, they demand higher nominal interest rates on their loans and investments. This results in a nominal interest rate that is comprised of two components: the real interest rate (the actual return on investment) and the expected inflation rate. When inflation rises, the expected inflation component of the nominal interest rate increases, even if the real return remains unchange An essential consideration in the context of Fisher's theory is the distinction between expected and unexpected inflation. Expected inflation refers to changes in the price level that are foreseen or anticipated by market participants. In contrast, unexpected inflation represents unanticipated fluctuations in inflation rates. The impact of these two forms of inflation on stock prices can be notably different. 
When inflation is expected and predictable, market participants can incorporate it into their decision-making processes. Investors, for instance, may adjust their investment strategies by seeking higher returns to offset the anticipated erosion of purchasing power. In such cases, stock prices may exhibit a negative response, as investors demand a higher real return on their investment Conversely, unexpected inflation introduces an element of surprise and uncertainty into the market. The sudden and unanticipated nature of inflation fluctuations can catch investors off guard. This uncertainty can lead to market volatility and affect stock prices negatively, as investors react to the unforeseen changes in the economic landscape.

The paper by Fama and Schwert (1977) estimates the extent to which various assets were hedges against the expected and unexpected components of the inflation rate during the 1953–1971 period. More specifically to my review, the paper delves into the intricate relationship between inflation and stock returns, with a particular focus on expected inflation. It is crucial to highlight that this study contributed significantly to the understanding of the link between economic indicators and stock market performance, especially the impact of inflation expectations. The methodology employed by Fama and Schwert is a cornerstone of their study. They conducted an empirical analysis using historical data on stock returns and inflation rates. Their unique contribution lies in the fact that they focused on expected inflation, differentiating their research from prior studies that considered only aggregate inflation rates. This methodology allowed them to delve deeper into the behavioral aspects of investors. The central finding of their study is the identification of a negative relationship between expected inflation and stock returns. In other words, when investors anticipate inflation, stock returns tend to be lower. This discovery carries substantial implications for investors and market analysts.

In their analysis of asset returns, the calculation of returns for a particular asset is based on changes in its price from time t-1 to t and the cash flows distributed to the asset's owners within that period, all relative to the asset's price at t-1. This calculation forms the foundation for assessing the performance of various assets. When it comes to common stocks, researchers frequently rely on two types of return data: equally-weighted portfolio returns, encompassing all New York Stock Exchange (NYSE) stocks, and value-weighted portfolio returns, which focus on selected NYSE stocks denoted as set and s, respectively. These return data series are typically obtained from the Center for Research in Security Prices at the University of Chicago. Notably, continuous compounding, the method of choice for calculating asset returns, is extensively applied to common stock returns. This approach offers a standardized means of assessing asset performance and is integral to many financial analyses.

Utilizing data spanning from January 1953 to July 1971, researchers conducted a regression analysis to estimate the relationship between the monthly return on a value-weighted portfolio of NYSE common stocks and the one-month treasury bill rate. The regression yielded the following equation: 


$$
s_t = 0.0234 - 5.5 * R_t + ε_t
$$

with an R-squared (R^2) of 0.03 and a standard error (S(ε)) of 0.0356. 

This equation provides an estimate of the expected nominal return on stocks for a given month (t) in relation to a proxy for the expected inflation rate (B_t), represented by the one-month bill rate set at the end of the preceding month (t - 1). In simplified terms, it suggests that the expected nominal stock return for a month is related to the expected inflation rate with an equation:

$$
E(5_t|B_t) = 0.0234 - 5.5 * B_t
$$

When the return on a one-month bill surpasses 0.0042, equivalent to 0.42 percent per month, the equation suggests a negative expected nominal return on the portfolio of NYSE common stocks. Notably, during a specific twenty-three-month period spanning from January 1969 to November 1970, the value of B_t consistently exceeded 0.0042. This consistency in B_t values indicates that expected nominal returns on NYSE stocks were consistently negative during this time frame. As an illustration, in February 1970, B_t measured 0.0063, implying an expected nominal return on stocks of approximately -1.17 percent for that particular month.

These findings hold significant importance. Regressions such as equation above highlight a consistent and dependable negative association between the anticipated returns on common stocks and the one-month treasury bill rate. Alternatively, considering that the coefficient of the interest rate in the equation is roughly three times the standard error below zero, it provides strong evidence that the risk premium associated with stocks, which represents the contrast between expected stock returns and bill returns, inversely fluctuates with changes in the interest rate.

Moving on to unexpected inflation, Ball and Cecchetti's (1990) study delved into the repercussions of unanticipated inflation. This paper delves into the intricate relationship between inflation and uncertainty, emphasizing the critical need to comprehend how inflation levels impact uncertainty. Although the common belief is that high inflation is linked to heightened uncertainty, empirical studies have yielded conflicting outcomes. To address this, the text distinguishes between short-term and long-term uncertainty in the context of inflation, revealing that high inflation predominantly affects long-term uncertainty. This discovery has significant implications. It clarifies the divergence in prior studies on inflation-uncertainty links, reinforces the idea that high inflation can destabilize monetary policy, and highlights the importance of maintaining low inflation to mitigate the costs of uncertainty, particularly in long-term contracts. The text also touches on the difference between inflation variability and uncertainty, emphasizing that high inflation increases both. Furthermore, it notes variations in the inflation-uncertainty relationship across countries and time periods. Ultimately, the passage provides a nuanced understanding of the multifaceted interplay between inflation and uncertainty.

The discussion revolves around various theories concerning the link between inflation and uncertainty, with a focus on scenarios where shifts in trend inflation are more likely to occur when the trend itself is at a high level. The author draws upon the insights of experts like Dennis Logue, Thomas Willett, and Milton Friedman to present a model. According to this model, when inflation is low, policymakers are generally expected to work towards keeping it at that level. Inflation might occur due to external shocks, but it's unlikely that policymakers would intentionally raise inflation. However, in the case of high trend inflation, the situation becomes less clear. Policymakers face a dilemma because they may want to reduce inflation but fear the potential recession that such an action could trigger. As a result, the timing of any disinflation is uncertain. The author formalizes this concept by suggesting that policymakers have varying opinions about the trade-offs between unemployment and inflation, affecting their preferences for disinflation. When inflation is high, the public becomes uncertain about future inflation because they cannot predict which policy views will ultimately prevail.
