Conversely, the element of surprise encapsulated in unexpected inflation ushers uncertainty into the market, instigating volatility and reshaping stock prices in unanticipated ways. ## Literature

In my paper, it is imperative to understand the pivotal works of Irving fisher to further clairfy why inflation impacts stock prices. The Fisher Effect posits a positive correlation between nominal interest rates and inflation rates. According to this theory, as inflation increases, nominal interest rates tend to rise correspondingly. This phenomenon can have profound implications for various aspects of the economy, including the stock market. {cite:t}`Fisher (1930)`s argument hinges on the idea that investors and lenders, when confronted with inflation, will seek to protect the purchasing power of their investments. To achieve this, they demand higher nominal interest rates on their loans and investments. This results in a nominal interest rate (denoted by *i*) that is comprised of two components: the real interest rate (denoted by *r*) and the expected inflation rate (denoted by *π*) in figure 1 below. When inflation rises, the expected inflation component of the nominal interest rate increases, even if the real return remains unchanged. This essentially makes figure 1 equation equivalent to figure 2 equation.

$$
(1 + i) = (1 + r) × (1 + π_{e})
$$

$$
i ≈ r + π_{e}
$$

An essential consideration in the context of Fisher's theory is the distinction between expected and unexpected inflation. Expected inflation refers to changes in the price level that are foreseen or anticipated by market participants. In contrast, unexpected inflation represents unanticipated fluctuations in inflation rates. The impact of these two forms of inflation on stock prices can be notably different:

1. Expected Inflation Component
   - If expected inflation (π)  increases, according to the Fisher equation, the nominal interest rate (i)  will increase, assuming the real interest rate remains constant.
   - Market participants can incorporate this into their decision-making processes. Investors, for instance, may adjust their investment strategies by seeking higher returns to offset the anticipated erosion of purchasing power. In such cases, stock prices may exhibit a negative response, as investors demand a higher real return on their investment.

2. Unexpected Inflation Component
   - When unexpected inflation occurs, it impacts the real interest   rate (r)  component of the Fisher equation. If nominal interest rates remain unchanged despite unexpected inflation, the real interest rate effectively changes. 
   - The sudden and unanticipated nature of inflation fluctuations can catch investors off guard. This uncertainty can lead to market volatility and affect stock prices negatively, as investors react to the unforeseen changes in the economic landscape.

It is important to note that both unexpected inflation and expected inflation negatively effect stock prices. This paper will delve into distinguishing the effect of expected vs unexpected inflation on stock prices further by citing historical research in multiple literature sources. 



{cite:t}`asset_fama_1977`'s pivotal work laid the groundwork for understanding the relationship between expected inflation and stock returns, employing a meticulous analysis of historical data to unveil a distinct negative correlation. By dissecting anticipated and unexpected inflation, their methodology showcased how investors demanded higher real returns when faced with expected inflation, resulting in diminished stock returns. Using data for the the January 1953 to July 1971 period, they ran a regression of the montly return to the value weighted portfolio of NYSE common stocks on the one month Treasury bill rate, which acts as a representation of the expected inflation component. The regression output was:

$$
s_{vt} = \underset{(0.0054)}{0.0234} - \underset{(1.85)}{5.50\beta_{1t}} + \hat{ε_{t}},              R_{adj}^2 = 0.03,             S(\hat{ε_{t}}) = 0.0356
$$

This simple OLS regression is easy to understand. If the expected inflation, represented by the one month Treasury bill yield, exceeds 0.0042, the figure 3 assesses a negative expected nominal return on the portfolio of NYSE common stocks. During the twenty-three month period from January 1969 through november 1970, the one month Treasury yield was always greater than 0.0042, which implies that the expected nominal returns on NYSE stocks were negative according to this relationship. As the paper measures expected inflation in terms of Treasury bill yields, the findings above can be stated more directly as a negative correlation between interest rates and stock returns. 

Building upon this foundation, {cite:t}`expected_domian_1996` strengthens this asymmetric association by incorporating time-series analysis over the period of January 1952 to December 1992. Stock returns were obtained from the Center for Research in Security Prices (CRSP) and the yields on one-month Treasury bills were obtained from the Ibbotson SBBI database. The research employs Augmented Dickey-Fuller tests to ensure stationarity of the stock return and interest rate change series, a prerequisite for the time series methods used in the study. It then proceeds to explore symmetric and asymmetric response models between stock returns and interest rate changes. The symmetric model reveals a consistent negative relationship between interest rate changes and stock returns across different interest rate series, aligning with earlier findings by Fama and Schwert (1981). However, the study extends its analysis to asymmetric relationships, separating positive and negative changes in interest rates. Results from the asymmetric response model indicate that the stock market primarily responds to interest rate decreases, with significant coefficients for negative interest rate changes and negligible coefficients for positive changes, suggesting an asymmetric effect on stock returns based on the direction of interest rate changes. The paper also examines the robustness of the findings by exploring asymmetries in stock market responses to changes in various interest rate series beyond the one-month Treasury bill data. Utilizing seven different interest rate series, including federal funds rates, three-month Treasury bills, and one-year to ten-year constant maturity data, the study investigates how stock returns react asymmetrically to changes in these rates. Notably, the most notable asymmetry is observed in the response to one-year maturities, where the market response demonstrates pronounced asymmetry. While the asymmetry is less distinct for three-year and five-year notes, it diminishes over the initial months for ten-year maturities. Similar patterns are found with longer-term Treasury bond yields. Additionally, the study examines excess stock returns relative to Treasury bill returns to control for nominal interest rate changes, finding that the asymmetry persists in these regressions as well. Overall, the analysis indicates consistent asymmetrical market responses to changes in various interest rate series, reinforcing the robustness of the observed asymmetry in stock market reactions.

Despite these consistent findings, {cite:p}`industry_boudoukh_1994` Bouboukh and Richardson's paper diverges from existing studies that predominantly focus on short-term asset returns with horizons of one year or less. Their research acknowledges a void in empirical literature concerning the application of the Fisher model across varying time horizons. While earlier studies concentrate on shorter durations, this study deliberately considers longer holding periods, recognizing that many investors maintain stocks over extended periods. Consequently, their paper investigates how stock returns behave in relation to inflation across these longer horizons. This departure from the prevailing research paradigm is crucial as it challenges the assumption that the Fisher model's implications remain consistent across all horizon lengths, signaling the importance of exploring inflation's impact on stocks over extended investment periods. The empirical analysis utilizes annual data on inflation, stock returns, and short-term/long-term interest rates spanning from 1802 to 1990. The paper uses Jeremy Siegel's data sources, detailed in his 1992a publication, which provide price level data closely resembling the Bureau of Labor Statistics' consumer price series. For yield data, Siegel employs long-term U.S. government bonds with maturities closest to 20 years, complemented by constructed one-year rates based on commercial paper rates, U.K. short-term rates (during the gold standard), and available U.S. government rates for the earlier sample period. The paper also uses the stock-return data, compiled by Schwert in 1990, which mirrors a market index. However, the paper notes that data before 1870, primarily consisting of financial firms and railways, is less comprehensive compared to later periods. Lastly, as there isn't a definitive definition of a "long" horizon, they opt for a somewhat arbitrary selection of five years. This choice yields nearly 40 independent observations within the overall sample, expanding the horizon considerably beyond previous studies.

$$
\sum \limits _{i=1} ^{5} \hat{R_{t+i}} = \hat{\alpha_{5}} + \hat{\beta_{5}}\sum \limits _{i=1} ^{5}π_{t+i} + \hat{ε_{t}}
$$

According to figure 4, the regression coefficient for the contemporaneous five-year stock returns against the five-year inflation rate shows a significantly positive relationship, standing at $\hat{\beta_{5}}$ = 0.52, with a standard error of 0.17. This finding suggests a tendency for nominal stock returns and inflation to move together within the sample, contradicting the works of Fama and Richardson on the short term affects of expected inflation to stock returns. Conversely, the estimate for $\hat{\beta_{1}}$, one year expected inflation, is near zero. This result aligns with prior research, which has consistently shown significantly negative estimates at shorter frequencies (monthly and quarterly) but estimates close to zero at the annual frequency. The hypothesis test comparing $\hat{\beta_{5}}$ = $\hat{\beta_{1}}$ against $\hat{\beta_{5}}$ > $\hat{\beta_{1}}$ yields a z statistic of 4.20, signifying significance at the 1-percent level. This divergence from previous literature is profoundly informative especially for investors who assess macroeconomic risks in terms of stock returns. According to Fama and Richardson, the short term effect (less than one year) of expected inflation on stock returns is negative; in contrast, the long term effect (more than five years) is positive. This information is to be acutely utilized for investors of different appetite, in terms of investment periods.

I wish to highlight the subtle difference regarding the goal of my paper and the literature written by the aforementioned scholars above. All the papers cited above analyze the effect of expected inflation on **nominal stock returns**. In my paper, I aim to research the effect of expected inflation in regards to **nominal stock prices**. Stock returns and stock prices are integral yet distinct aspects of evaluating investment performance in the financial markets. A stock market return encapsulates the overall change, whether positive or negative, in the value of an investment or asset over a specified period. When an investment generates a positive return, it signifies a profit accrued during that period, reflecting an increase in the investment's value. Conversely, a negative return denotes a loss incurred on the investment, indicative of a decrease in its value over the same timeframe. "Total returns" encompass various components beyond solely the change in stock price, integrating dividends and interest payments alongside the stock's price fluctuation. On the other hand, the term "nominal return" specifically refers to the change in stock price alone, excluding dividends and interest payments from calculation. Understanding this distinction is crucial in investment analysis. Stock prices signify the current value of a company's shares in the market, reflecting supply and demand dynamics, company performance, and broader economic factors. In contrast, stock returns offer a more investor oriented perspective, accounting for both price changes and additional income generated from dividends and interest payments. As such, I tried to capture the effect of expected inflation in terms of the more broader market perspective.

Moving on to unexpected inflation, {cite:p}`inflation_ball_1990` Ball and Cecchetti's (1990) study delved into the repercussions of unanticipated inflation. This paper delves into the intricate relationship between inflation and uncertainty, emphasizing the critical need to comprehend how inflation levels impact uncertainty. Although the common belief is that high inflation is linked to heightened uncertainty, empirical studies have yielded conflicting outcomes. To address this, the text distinguishes between short-term and long-term uncertainty in the context of inflation, revealing that high inflation predominantly affects long-term uncertainty. This discovery has significant implications. It clarifies the divergence in prior studies on inflation-uncertainty links, reinforces the idea that high inflation can destabilize monetary policy, and highlights the importance of maintaining low inflation to mitigate the costs of uncertainty, particularly in long-term contracts. The text also touches on the difference between inflation variability and uncertainty, emphasizing that high inflation increases both. Furthermore, it notes variations in the inflation-uncertainty relationship across countries and time periods. Ultimately, the passage provides a nuanced understanding of the multifaceted interplay between inflation and uncertainty.

The discussion revolves around various theories concerning the link between inflation and uncertainty, with a focus on scenarios where shifts in trend inflation are more likely to occur when the trend itself is at a high level. The author draws upon the insights of experts like Dennis Logue, Thomas Willett, and Milton Friedman to present a model. According to this model, when inflation is low, policymakers are generally expected to work towards keeping it at that level. Inflation might occur due to external shocks, but it's unlikely that policymakers would intentionally raise inflation. However, in the case of high trend inflation, the situation becomes less clear. Policymakers face a dilemma because they may want to reduce inflation but fear the potential recession that such an action could trigger. As a result, the timing of any disinflation is uncertain. The author formalizes this concept by suggesting that policymakers have varying opinions about the trade-offs between unemployment and inflation, affecting their preferences for disinflation. When inflation is high, the public becomes uncertain about future inflation because they cannot predict which policy views will ultimately prevail.
